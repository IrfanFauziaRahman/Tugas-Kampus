```python
import math
import matplotlib.pyplot as plt
```

```python
# variabel
ke = 8.99 * 10**9  # konstanta Coulomb dalam N·m²/C²
q = 1.00 * 10**-6  # muatan Coulomb (C)

# Hitung nilai medan listrik E dan fluks listrik Φ_E untuk r = 1.00 meter
r = 1.00  # radius meter (m)
E = ke * (q / r**2)
Phi_E = E * 4 * math.pi * r**2

# Tampil nilai medan listrik dan fluks listrik
E_str = f"{E:.2e}"
E_formatted = E_str.replace("e+03", " × 10³")
print(f"Medan Listrik, E = {E_str} N/C atau {E_formatted} N/C")

Phi_E_str = f"{Phi_E:.2e}"
Phi_E_formatted = Phi_E_str.replace("e+05", " × 10⁵")
print(f"Fluks Listrik, Φ_E = {Phi_E_str} N·m²/C atau {Phi_E_formatted} N·m²/C")
```

```python
r_values = [0.1 * i for i in range(1, 21)]  # radius meter (m)
E_values = []
Phi_E_values = []

# Hitung medan listrik E dan fluks listrik Φ_E setiap radius
for r in r_values:
    E = ke * (q / r**2)
    A = 4 * math.pi * r**2 
    Phi_E = E * A
    
    E_values.append(E)
    Phi_E_values.append(Phi_E)

# Plot hasil
plt.figure(figsize=(12, 6))

# Grafik Medan Listrik
plt.subplot(1, 2, 1)
plt.plot(r_values, E_values, marker='o', color='b')
plt.title('Medan Listrik vs Jarak')
plt.xlabel('Jarak (m)')
plt.ylabel('Medan Listrik (N/C)')
plt.grid(True)
plt.axhline(y=E, color='r', linestyle='--', label=f'E = {E_formatted} N/C')
plt.legend()

# Grafik Fluks Listrik
plt.subplot(1, 2, 2)
plt.plot(r_values, Phi_E_values, marker='o', color='r')
plt.title('Fluks Listrik vs Jarak')
plt.xlabel('Jarak (m)')
plt.ylabel('Fluks Listrik (N·m²/C)')
plt.grid(True)
plt.axhline(y=Phi_E, color='b', linestyle='--', label=f'Φ_E = {Phi_E_formatted} N·m²/C')
plt.legend()

plt.tight_layout()
plt.show()
```
